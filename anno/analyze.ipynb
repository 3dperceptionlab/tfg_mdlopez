```python
# read biovid_ds_train.csv, biovid_ds_val.csv, biovid_ds_test.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# read data
train_df = pd.read_csv('biovid_ds_train.csv', sep=' ', header=None)
val_df = pd.read_csv('biovid_ds_val.csv', sep=' ', header=None)
test_df = pd.read_csv('biovid_ds_test.csv', sep=' ', header=None)

```

```python
# create a function to plot the distribution of labels
def plot_distribution(df, title):
    df[1].value_counts().plot(kind='bar')
    plt.title(title)
    plt.xlabel('Clase')
    plt.ylabel('NÃºmero de muestras')
    plt.show()
```

```python
# plot the distribution of labels
plot_distribution(train_df, 'DistribuciÃ³n de clases en el conjunto de entrenamiento')
```

```python
# val_df
plot_distribution(val_df, 'DistribuciÃ³n de clases en el conjunto de validaciÃ³n')
```

```python
# test
plot_distribution(test_df, 'DistribuciÃ³n de clases en el conjunto de test')
```
